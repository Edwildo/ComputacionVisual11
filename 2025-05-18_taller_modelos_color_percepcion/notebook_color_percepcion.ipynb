```python
import cv2
import matplotlib.pyplot as plt
from skimage import color

img = cv2.imread('Ducati.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
img_lab = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2Lab)
```

```python
def show_channels(image, space_name):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    titles = [f"{space_name} - Channel {i+1}" for i in range(3)]
    for i in range(3):
        axs[i].imshow(image[:, :, i], cmap='gray')
        axs[i].set_title(titles[i])
        axs[i].axis('off')
    plt.show()

show_channels(img_hsv, "HSV")
show_channels(img_lab, "Lab")
```

```python
def simulate_low_light(image, brightness_factor=0.3):
    return cv2.convertScaleAbs(image, alpha=brightness_factor, beta=0)

low_light = simulate_low_light(img_rgb)
plt.imshow(low_light)
plt.title("Simulacin de Baja Iluminacin")
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

def switch_color_model(image, mode='hsv'):
    if mode == 'hsv':
        return cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    elif mode == 'lab':
        return cv2.cvtColor(image, cv2.COLOR_RGB2Lab)
    return image

# 1. Cargar imagen y convertir a RGB
img_bgr = cv2.imread('Ducati.jpg')  # Reemplaza con la ruta correcta
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# 2. Aplicar conversin a HSV o Lab
img_hsv = switch_color_model(img_rgb, mode='hsv')

# 3. Mostrar la imagen convertida canal por canal
def show_channels(image, title_prefix):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    for i in range(3):
        axs[i].imshow(image[:, :, i], cmap='gray')
        axs[i].set_title(f'{title_prefix} - Canal {i+1}')
        axs[i].axis('off')
    plt.tight_layout()
    plt.show()

show_channels(img_hsv, 'HSV')
```
